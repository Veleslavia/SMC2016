```python
import pandas as pd
import pickle
from sklearn import preprocessing
from sklearn.metrics import precision_recall_fscore_support

data = pd.DataFrame.from_csv("../audio/features_audio_video.csv")
data = data.dropna(axis=1, how='any')

y = data['class']

X = data.drop(['class'], axis=1)
X.shape
```

```python
clf = pickle.load(open("irmas_xgb_classifier.plk"))
le = pickle.load(open("irmas_le.plk"))
inputer = pickle.load(open("rwc_inputer.plk"))
le2 = pickle.load(open("irmas_le2.plk"))

X = data.drop(['class'], axis=1).values
X = inputer.transform(X)

y_predict = clf.predict(X)
y_predict_proba = clf.predict_proba(X)
```

```python
le.inverse_transform(y_predict)
```

```python
import heapq

top2 = list()
for i, elem in enumerate(y_predict_proba):
    pred = heapq.nlargest(2, range(len(elem)), elem.__getitem__)    
    lst = [le2.classes_[idx] + ": " + str(round(elem[idx], 4)) for idx in pred]
    top2.append(lst)
```
